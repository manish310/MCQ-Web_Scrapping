```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
for i in range(1,20):
    gk="https://www.gktoday.in/quizbase/general-science-chemistry?pageno=" + str(i)
    #print(gk)
    page=urllib.request.urlopen(gk)
    soup=BeautifulSoup(page)
    all_ques=soup.find_all('div',class_=['wp_quiz_question testclass','wp_quiz_question_options','ques_answer'])
    for ques in all_ques:
        print(ques.get_text())
```
